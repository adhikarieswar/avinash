```python
import neattext.functions as nfx
```

```python
data['course_title']=data['course_title'].apply(nfx.remove_stopwords)
# stop words like an,in, is, was are removed out
```

```python
data['course_title']=data['course_title'].apply(nfx.remove_special_characters)
```

```python
data.columns
```

```python




# count vectorizer converts text data to numercal vectors
# bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=3000)
```

```python
vector=cv.fit_transform(data['course_title']).toarray()
vector
```

```python


# cosine similarity
# cosine similarity is a metric used to measure similarity between two vectors  is  a multi dimensional space
# cosine similarity measures the similarity between two vectors by calculating the cosine of the angle between the two vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(vector)
```

```python


cs # cosine similarity scores based on cos theta=a.b/mod(a).mod(b)
```

```python
cs.shape
```

```python
cs[0]
```

```python
sorted(enumerate(cs[0]),reverse=True,key=lambda x:x[1])[1:6]
```
