```python
# creating function  in order to load  the and
def load_data(path):
  input_file=path
  with open(input_file,'r') as f:
    data=f.read()
    return data.split("\n")



```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
french_tokenizer=Tokenizer()
english_tokenizer=Tokenizer()
```

```python
_

english_tokenizer.fit_on_texts(english_sentence)
french_data_numeric=french_tokenizer.fit_on_texts(french_sentence)
french_tokenizer.word_index

```

```python
english_text_numeric=english_tokenizer.texts_to_sequences(english_sentence)

french_text_numeric=french_tokenizer.texts_to_sequences(french_sentence)
french_text_numeric
```

```python

# pad sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python



 max_english_length=max(len(text) for text in english_text_numeric)
 max_french_length=max(len(text) for text in  french_text_numeric)
max_english_length
max_french_length
  # after finding length of all th lists one after another then max function is applied to it
```

```python
# pad_sequences function is applied to the  text_numeric  in order to  avoid deficit where  values are  less compared with
# max_sequence_length   those  can be replaced with zeros  but padding='post' gives the values that should be inserted should be at the end of sequences
encoder_input_data=pad_sequences(english_text_numeric,maxlen=max_english_length,padding='post')
decoder_input_data=pad_sequences(french_text_numeric,maxlen=max_french_length,padding='post')
decoder_output_data=pad_sequences(french_text_numeric[1:],maxlen=max_french_length,padding='post')
decoder_output_data.shape
encoder_input_data=encoder_input_data[0:137860]
encoder_input_data.shape









```

```python

from tensorflow.keras.models import Model
from keras.layers  import TimeDistributed
```

```python
# building model of  LSTM BASED NMT  MODELS
from tensorflow.keras.layers import   Dense,LSTM,Embedding,Input
embedding_size=64
max_english_length=15
max_french_length=21
hidden_size=128
english_vocab_size=len(english_tokenizer.word_index)+1
french_vocab_size=len(french_tokenizer.word_index)+1
# encoder part
encoder_input=Input(shape=(max_english_length,))
encoder_embedding=Embedding(english_vocab_size,embedding_size)(encoder_input)
lstm_model,state_h,state_c=LSTM(hidden_size,return_state=True)(encoder_embedding)
encoder_states=[state_h,state_c]
# Decoder part
decoder_input=Input(shape=(max_french_length,))
decoder_embedding=Embedding(french_vocab_size,embedding_size)(decoder_input)
decoder_lstm, _, _=LSTM(hidden_size,return_sequences=True,return_state=True)(decoder_embedding)
decoder_dense=TimeDistributed(Dense(french_vocab_size ,activation='softmax'))
decoder_output=decoder_dense(decoder_lstm)
# CREATING INSTANCE OF MODEL
model=Model([encoder_input,decoder_input],decoder_output)
model.compile(optimizer='Adam',loss='SparseCategoricalCrossentropy',metrics=['accuracy'])
model.summary()























```

```python

# we are going to train the model here

decoder_output.shape
model_training=model.fit([encoder_input_data,decoder_input_data],decoder_output_data,batch_size=60,epochs=60,validation_split=0.2)

decoder_output_data.shape
decoder_input_data.shape



```

```python
val_loss=model_training.history['val_loss']
val_accuracy=model_training.history['val_accuracy']
```

```python
best_val_loss=min(val_loss)
best_val_accuracy=max(val_accuracy)
```

```python
epoch_of_bestloss=val_loss.index(best_val_loss)+1
epoch_of_bestaccuracy=val_accuracy.index(best_val_accuracy)+1
epoch_of_bestaccuracy
```

```python
# we will consider best loss means in each epoch with different hyper parameters parameters we are training model in that case  at  epoch  7 we are going to observe best loss  minimum loss
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
es=EarlyStopping(monitor="val_loss")
mcp=ModelCheckpoint(filepath="best_model.keras",monitor="val_loss",save_best_only=True)
model.fit([encoder_input_data,decoder_input_data],decoder_output_data,batch_size=60,epochs=60,validation_split=0.2,callbacks=[es,mcp])


```

```python
# Dual Language Translator
import ipywidgets as widgets
from IPython.display import display
from transformers import MarianMTModel, MarianTokenizer

# Load translation models
tokenizer_fr = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
model_fr = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
tokenizer_hi = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model_hi = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translate(change):
    input_text = input_box.value
    if len(input_text) < 10:
        result_label.value = "Upload again. Word must be 10 or more letters."
    else:
        # Translate to French
        encoded_input_fr = tokenizer_fr.encode(input_text, return_tensors="pt")
        translated_fr = model_fr.generate(encoded_input_fr)
        french = tokenizer_fr.decode(translated_fr[0], skip_special_tokens=True)

        # Translate to Hindi
        encoded_input_hi = tokenizer_hi.encode(input_text, return_tensors="pt")
        translated_hi = model_hi.generate(encoded_input_hi)
        hindi = tokenizer_hi.decode(translated_hi[0], skip_special_tokens=True)

        # Update result
        result_label.value = f"French: {french}\nHindi: {hindi}"

# Create widgets for input and output
input_box = widgets.Text(
    description='Enter Text:',
    placeholder='Type something here'
)
result_label = widgets.Label(value="Translation will appear here.")

# Create a button to trigger translation
translate_button = widgets.Button(description="Translate")
translate_button.on_click(translate)

# Display the widgets
display(input_box, translate_button, result_label)
```
